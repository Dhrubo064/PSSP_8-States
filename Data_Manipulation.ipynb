**MainSet_Before Extraction**

```python
# Calculate 8-state (SS8) class percentages from a CSV that has an 'sst8' column.
# Works in Google Colab. If needed, change PATH to your file location in Colab.

import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

# 1) Point to your dataset
PATH = "/content/2022-08-03-ss.cleaned.csv"   # <-- change if needed (your local/Drive path)

# 2) Load the dataset
df = pd.read_csv(PATH, low_memory=False)

# 3) Count residues per SS8 class by expanding the per-row 'sst8' strings
seq_col = "sst8"  # your dataset's column with 8-state labels
counter = Counter()

for s in df[seq_col].dropna().astype(str):
    s = s.strip().replace(" ", "")
    for ch in s:
        counter[ch] += 1

# 4) Build results table
res = pd.DataFrame.from_dict(counter, orient="index", columns=["count"]).sort_index()
total = res["count"].sum()
res["percentage"] = (res["count"] / total * 100.0).round(6)

# (Optional) order by conventional SS8 label order
order = ["H", "G", "I", "E", "B", "T", "S", "C", "-"]
ordered_index = [x for x in order if x in res.index] + [x for x in res.index if x not in order]
res = res.loc[ordered_index]

# 5) Show the table
print(res.rename_axis("class"))

# 6) Save outputs
# res.to_csv("ss8_class_distribution.csv")
# print("Saved table to ss8_class_distribution.csv")

# 7) Plot a bar chart (Matplotlib, single plot, default colors)
plt.figure()
ax = res["percentage"].plot(kind="bar")
ax.set_xlabel("Class")
ax.set_ylabel("Percentage")
ax.set_title("SS8 Class Distribution (%)")
plt.tight_layout()
plt.show()
```
